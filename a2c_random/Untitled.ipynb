```python
import pickle
log_dict = None
with open('log_a2c_random.pickle', 'rb') as f:
    log_dict = pickle.load(f)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
hits_diff = log_dict["mean_hits_diff"]
hits_diff -= min(hits_diff)
hits_diff /= max(hits_diff)
plt.figure(figsize=(7,4))
plt.xlabel('Epoch')
plt.grid()
plt.plot(range(len(log_dict["mean_win"])), log_dict["mean_win"], label="Mean win rate")
plt.plot(range(len(log_dict["mean_hits_diff"])), hits_diff, label="Health difference (normalized)")
plt.legend()
plt.title('Win rate & health difference correlation')
```
